```
from keras.preprocessing import image
from keras.utils import np_utils
from sklearn.utils import shuffle
```

```
p = Path("./train/")
dirs = p.glob("*")

image_data = []
labels = []

image_paths = []

for folder_dir in dirs:
    label = str(folder_dir).split("\\")[-1]
    
    cnt = 0
    print(folder_dir)
    
    #Iterate over all folder_dir and pick all images of coins
    for img_path in folder_dir.glob("*.jpg"):
        img = image.load_img(img_path,target_size=(224,224))
        img_array = image.img_to_array(img)
        # img_array = img_array/255
        image_data.append(img_array)
        labels.append(label)
        cnt+=1
        
    print(cnt)
```

```
print(len(image_data),len(labels))
```

```
import random
```

```
combined = list(zip(image_data,labels))
random.shuffle(combined)
```

```
image_data[:],labels[:] = zip(*combined)
```

```
print(labels[0][6:])
```

```
new_label = []
```

```
for i in range(len(labels)):
  new_label.append(labels[i][6:])
```

```
print(new_label[:5])
```

```
new_label = list(map(int, new_label))
```

```
print(new_label[:5])
```

```
import numpy as np
```

```
X_train = np.array(image_data)
Y_train = np.array(new_label)
print(X_train.shape)
print(Y_train.shape)
```

```
from keras.utils import np_utils
```

```
Y_train = np_utils.to_categorical(Y_train)
print(X_train.shape)
print(Y_train.shape)
```

```
Y_train = Y_train[:,1:]
print(X_train.shape)
print(Y_train.shape)
```

Transfer **Learning**

```
from keras.applications.resnet50 import ResNet50
from keras.optimizers import Adam
from keras.models import Model
from keras.layers import *
```

```
model = ResNet50(include_top=False,weights='imagenet',input_shape=(224,224,3))
```

```
model.summary()
```

```
av1 = GlobalAveragePooling2D()(model.output)
fc1 = Dense(256,activation='relu')(av1)
d1  = Dropout(0.5)(fc1)
fc2 = Dense(211,activation='softmax')(d1)
```

```
model_new = Model(input=model.input,outputs=fc2)
model_new.summary()
```

```
adam = Adam(lr=0.0003)
```

```
model_new.compile(loss='categorical_crossentropy',optimizer=adam,metrics=['accuracy'])
```

```
for ix in range(len(model_new.layers)):
  print(ix,model_new.layers[ix])
```

```
for ix in range(169):
  model_new.layers[ix].trainable=False

model_new.compile(loss='categorical_crossentropy',optimizer=adam,metrics=['accuracy'])
model_new.summary()
```

```
hist = model_new.fit(
    X_train,
    Y_train,
    shuffle = True,
    batch_size = 16,
    epochs = 5,
    validation_split=0.20
)
```
